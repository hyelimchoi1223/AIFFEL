### 선 그래프 그리기

```python
from datetime import datetime
import pandas as pd
import os

# 그래프 데이터
csv_path = os.getenv("HOME")+"/aiffel/data_visualization/data/AMZN.csv"
data = pd.read_csv(csv_path, index_col=0, parse_dates=True)
price = data['Close']

# 축 그리기 및 좌표축 설정
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
price.plot(ax=ax, style='black') # Pandas의 Plot을 사용. 위에서 정의한 ax를 사용했다.

# x, y축 설정
plt.ylim([1600, 2200])
plt.xlim(['2019-05-01', '2020-03-01'])

# important_data에 나오는 부분을 표시하기 위한 코드
# 주석을 다는 메서드 = annotate()
important_data = [(datetime(2019, 6, 3), "Low Price"), (datetime(2020,2,19), "Peak Price")]
for d, label in important_data:
    ax.annotate(label, xy=(d,price.asof(d)+10),
               xytext=(d,price.asof(d)+100),
               arrowprops=dict(facecolor='red'))
# 그리드 : 격자 눈금 추가
plt.grid()
ax.set_title('StockPrice') # =plt.title('StockPrice')

# plt.savefig('./sample.png') 그래프를 이미지로 출력
plt.show()
```

```python
# 서브플롯 생성
x = np.linspace(0,10,100)

plt.subplot(2,1,1)
plt.plot(x, np.sin(x), 'orange', 'o')

plt.subplot(2,1,2)
plt.plot(x, np.cos(x), 'black', '--')

plt.show()
```

```python
# linestyle, marker
# 라인 스타일은 plot 인자로 들어간다.
x = np.linspace(0,10,100)

plt.plot(x, x + 0, linestyle='solid') 
plt.plot(x, x + 1, linestyle='dashed') 
plt.plot(x, x + 2, linestyle='dashdot') 
plt.plot(x, x + 3, linestyle='dotted')
plt.plot(x, x + 0, '-g') # solid green 
plt.plot(x, x + 1, '--c') # dashed cyan 
plt.plot(x, x + 2, '-.k') # dashdot black 
plt.plot(x, x + 3, ':r'); # dotted red
plt.plot(x, x + 4, linestyle='-') # solid 
plt.plot(x, x + 5, linestyle='--') # dashed 
plt.plot(x, x + 6, linestyle='-.') # dashdot 
plt.plot(x, x + 7, linestyle=':'); # dotted
```

```python
# 히스토그램 예제
# 히스토그램 : 도수분포표를 그래프로 나타낸것
# 도수분포표 : 자료의 분포 상태를 의미한다.
# 가로 : 계급->변수 구간
# 세로 : 도수-> 빈도수

mu1, mu2, sigma = 100,130,15 # mu1 = x1평균, mu2 = x2평균, sigma = 표준편차
x1 = mu1 + sigma * np.random.randn(10000) # randn : 표준 정규분포에서 난수 생성
x2 = mu2 + sigma * np.random.randn(10000)

# 축 그리기
fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)

# 그래프 그리기
patches = ax1.hist(x1, bins=50, density=False) # bins는 x를 50개 구간으로 나눈다.
patches = ax1.hist(x2, bins=50, density=False, alpha=0.5)
ax1.xaxis.set_ticks_position('bottom') # x축 눈금 표시
ax1.yaxis.set_ticks_position('left')   # y축 눈금 표시

plt.xlabel('Bins')
plt.ylabel('Numver of Values in Bin')
ax1.set_title('Two Frequency Distributions')

plt.show()
```

### Heatmap
2차원으로 시각화하여 보여주는 것. 수치에 따른 색상으로 나타냄

##### pivot
데이터를 축을 기준으로 바꾼다는 얘기이다.
column->index, index->column 이런 식으로

```python
pivot = flights.pivot(index='year', columns='month', values='passengers')
pivot
```

```python
sns.heatmap(pivot)
```

```python
# 다양한 옵션도 있다.
sns.heatmap(pivot, linewidths=.2, annot=True, fmt="d")
# annot : 숫자를 표시할지 말지 옵션
# fmt="d" : 데이터 포맷
```

```python
# 다른 색깔로 표시하기
sns.heatmap(pivot, cmap="YlGnBu")
```
